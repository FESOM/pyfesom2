# Remote datasets 

<p style="color:gray;font-size:18px;">Pyfesom2 comes with remote datasets that are loaded on demand.</p>

```python
from pyfesom2 import datasets
```

## 1. Tutorial datasets

<p style="color:gray;font-size:18px;"> A smallest and quick to load dataset </p>

```python
datasets.tutorial_dataset.load()
```

<p style="color:gray;font-size:18px;">A CMIP6-like and a more comprehensive dataset.</p>

```python
datasets.core.load()
```

## 2. Frontier simulation examples

<p style="color:gray;font-size:18px;">Pyfesom2 comes with examples of frontier simulations</p>

#### A sample Arctic 1km resolution dataset with ~11 million surface nodes.

```python
datasets.arctic_1km.load()
```

#### A sample Rossby 4.2 simulation with ~23 million surface nodes and 79 levels.

```python
datasets.rossby42.load()
```

#### As these remote datasets are currently statically served there are other versions of this rossby4.2 that are optimized level: datasets.rossby42_level and spatial: datasets.rossby42_spatial based access.

## 3. AWI-CM CMIP6 grids

<p style="color:gray;font-size:18px;">These datasets are just coordinates corresponding to LR, MR, HR versions of AWI-CM model used in CMIP6. Currently these contain face information that
contain missing information in ESGF and necessary for spatial plotting. Soon using diagnostics from pyfesom2 may be supported on these datasets. </p>

<p style="color:gray;font-size:18px;">
Merging this grid data with CMIP6, AWI-CM can be done using xarray.merge([grid_data, cmip6_data]).</p>

```python
datasets.cmip6_hr.load()
```

#### Use datasets.cmip6_lr.load() and datasets.cmip6_mr.load() for LR and MR respectively 

## 4. FESOM-like dataset 

<p style="color:gray;font-size:18px;"> Most useful for developers to test, benchmark new methods on FESOM-like grids, that may not be available locally. If dask is available then a data variable is returned along else only coordinate dataset is returned. Check method doc for more use cases.</p>

```python
datasets.fesom_like(100000)
```

```python
datasets.fesom_like(100000, times=10, levels=10)
```
